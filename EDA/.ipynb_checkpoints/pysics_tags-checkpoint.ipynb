```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
tags_df = pd.read_csv("physics_csv/Tags.csv")
```

```python
tags_df["Count"].describe()
```

```python
plt.figure(figsize=(10, 5))
plt.hist(tags_df["Count"], bins=50, log=True)

plt.title("Distribution of Tag Usage Counts (Log Scale)")
plt.xlabel("Tag Usage Count")
plt.ylabel("Number of Tags")


plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda y, _: f'{int(y)}'))

plt.tight_layout()
plt.show()
```

```python
## word cloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt

tag_freq = dict(zip(tags_df['TagName'], tags_df['Count']))

wordcloud = WordCloud(width=1000, height=500, background_color='white').generate_from_frequencies(tag_freq)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Tag Frequency Word Cloud", fontsize=18)
plt.show()
```

```python
####### 
print("how many tags：", tags_df.shape[0])

# Top 10 
top_tags = tags_df.sort_values("Count", ascending=False).head(10)
print(top_tags[["TagName", "Count"]])
```

```python
# hist
plt.figure(figsize=(10,6))
plt.bar(top_tags["TagName"], top_tags["Count"])
plt.xticks(rotation=60)
plt.title("Top 10 Most Common Tags")
plt.ylabel("Usage Count")
plt.tight_layout()
plt.show()
```

```python
#################feature overview#################
# Preview
print("Head of the table:")
print(tags_df.head(), end="\n\n")

# Data types
print("Data Types:")
print(tags_df.dtypes, end="\n\n")

# Missing values
print("Missing Values:")
print(tags_df.isnull().sum(), end="\n\n")

# Unique values
print("Unique Value Counts:")
for col in tags_df.columns:
    print(f"{col}: {tags_df[col].nunique()} unique values")
```

```python
# Distribution checks (e.g., Count column)
plt.figure(figsize=(10, 5))
plt.hist(tags_df['Count'], bins=50, log=True)
plt.title("Distribution of Tag Usage Count")
plt.xlabel("Count")
plt.ylabel("Frequency (log scale)")
plt.tight_layout()
plt.show()
```
