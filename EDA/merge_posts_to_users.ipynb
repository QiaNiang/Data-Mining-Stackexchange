```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import re
import seaborn as sns
import matplotlib.dates as mdates
from datetime import datetime
```

```python
folder = r"D:\Studium\DM-Dataset\math.stackexchange.com\CSV"
file1 = "math_sentiment_results.csv"
path1 = os.path.join(folder, file1)
df_posts = pd.read_csv(path1)

file2 = "Users.csv" 
path2 = os.path.join(folder, file2)
df_users = pd.read_csv(path2)
```

```python
df_posts.info()
```

```python
df_users.info()
```

```python
df_posts["OwnerUserId"] = df_posts["OwnerUserId"].astype("Int64") 
df_users["Id"] = df_users["Id"].astype("Int64")
```

```python
merged_df = df_posts.merge(
    df_users,
    how="right",
    left_on="OwnerUserId",
    right_on="Id",
    suffixes=('_post', '_user')
)
```

```python

id_count_columns = [col for col in merged_df.columns if 'id' in col.lower() or 'count' in col.lower()]

for col in id_count_columns:
    merged_df[col] = merged_df[col].astype("Int64")

```

```python
merged_df
```

```python
filename = "posts_to_users.csv"
output_path = f"{folder}\\{filename}"

merged_df.to_csv(output_path, index=False, encoding='utf-8-sig')
```
