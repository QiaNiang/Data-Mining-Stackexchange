```python
# Users who have non-null comments
has_comment = df["comments.Text"].notna()

# Users who have non-null posts (we can use 'posts.Title' or 'posts.SentimentScore')
has_post = df["posts.Title"].notna() | df["posts.SentimentScore"].notna()

# Combined condition
has_comment_or_post = has_comment | has_post

# Check if any user-row violates this
if not has_comment_or_post.all():
    print("‚ùå Some users were included who had neither posts nor comments.")
    display(df[~has_comment_or_post])
else:
    print("‚úÖ All users have either a comment, a post, or both.")
```

```python
# Users with only comments
comment_users = df[df["comments.Text"].notna()]["UserId"].unique()
post_users = df[df["posts.Title"].notna() | df["posts.SentimentScore"].notna()]["UserId"].unique()

only_comment_users = set(comment_users) - set(post_users)
only_post_users = set(post_users) - set(comment_users)

print(f"üë• Users with ONLY comments: {len(only_comment_users)}")
print(f"üìù Users with ONLY posts: {len(only_post_users)}")

# Show example rows for a user with only comments
if only_comment_users:
    example_comment_user = list(only_comment_users)[0]
    print(f"\nüîç Example user with only comments: {example_comment_user}")
    display(df[df["UserId"] == example_comment_user])

# Show example rows for a user with only posts
if only_post_users:
    example_post_user = list(only_post_users)[0]
    print(f"\nüîç Example user with only posts: {example_post_user}")
    display(df[df["UserId"] == example_post_user])
    
```

```python
# Normalize nulls and empty strings
comments_clean = df["comments.Text"].fillna("").str.strip()
posts_title_clean = df["posts.Title"].fillna("").str.strip()
posts_sentiment_exists = df["posts.SentimentScore"].notna()

# Check conditions
has_comment = comments_clean != ""
has_post = (posts_title_clean != "") | posts_sentiment_exists

# Final check
has_comment_or_post = has_comment | has_post

# Inspect failures
if not has_comment_or_post.all():
    print("‚ùå Some rows are missing both comment and post info.")
    invalid_rows = df[~has_comment_or_post]
    print(f"üõë Invalid row count: {len(invalid_rows)}")
    display(invalid_rows)
else:
    print("‚úÖ All rows include at least a post or a comment.")
print(df.columns.tolist())
```

```python
import pandas as pd

# Load merged dataset
merged_path = "/Users/adilshamji/Documents/25-Data-mining-lab/Data-Mining-Stackexchange/data/users_with_posts_and_or_comments.csv"
df = pd.read_csv(merged_path)

# Check for CleanBodyNoMath column
if "CleanBodyNoMath" not in df.columns:
    print("‚ùå 'CleanBodyNoMath' column is missing.")
else:
    # Clean and filter non-empty values
    df["CleanBodyNoMath"] = df["CleanBodyNoMath"].fillna("").astype(str)
    non_empty = df[df["CleanBodyNoMath"].str.strip() != ""]

    # Summary stats
    print(f"‚úÖ Total rows: {len(df)}")
    print(f"üßæ Non-empty CleanBodyNoMath rows: {len(non_empty)} ({len(non_empty)/len(df)*100:.2f}%)")

    # Sample preview
    print("\nüìù Sample CleanBodyNoMath values:")

    # Optional: length analysis
    non_empty["body_length"] = non_empty["CleanBodyNoMath"].str.len()
    print("\nüìä Length stats of CleanBodyNoMath:")
    print(non_empty["body_length"].describe())
```

```python
# 1. All post IDs that exist
all_post_ids = set(posts_df["PostId"])

# 2. Post IDs that have at least one comment
commented_post_ids = set(comments_df["PostId"].dropna())

# 3. Posts without comments
uncommented_post_ids = all_post_ids - commented_post_ids

# 4. Filter the posts
posts_without_comments = posts_df[posts_df["PostId"].isin(uncommented_post_ids)]

# Optional: split into questions and answers
questions_no_comments = posts_without_comments[posts_without_comments["PostTypeId"] == "1"]
answers_no_comments   = posts_without_comments[posts_without_comments["PostTypeId"] == "2"]

# Print results
print(f"üìå Total posts without comments: {len(posts_without_comments)}")
print(f"‚ùì Questions without comments: {len(questions_no_comments)}")
print(f"üí¨ Answers without comments: {len(answers_no_comments)}")

# Optional preview
display(posts_without_comments.head())
```
